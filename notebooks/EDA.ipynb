```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Import data file.

df = pd.read_csv(r"C:\Users\hp\My Project\Project Cat Breeds v2 (Classification) (10 class)\cat_breeds_clean.csv")

df
```

```python
#Make sure the names are clear and understandable in each column.

print(df.columns)
```

```python
#Data summary

df.info()
```

```python
#Ensure that each column has a logical data type and understandable values.

print(df.head())
```

```python
#Analyze unique values for each column.

for col in df.columns:
    print(f"{col} - number of nunique value :", df[col].nunique())
```

```python
#Are there any columns that contain illogical values (such as negative numbers for age or weight).

print(df[df['Weight'] < 0])
print(df[df['length'] < 0])
print(df[df['Age'] < 0])
```

```python
#Check balanced distribution (Percentages for each category).

print(round(df['Breeds'].value_counts(normalize=True) * 100, 1))
```

```python
#A graph showing the distribution.

df['Breeds'].value_counts().plot(kind='bar')
plt.title('Distribution of Cat Breeds')
plt.xlabel('Breed')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Age and weight distribution plot for each breed.

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Breeds', y='Age')
plt.xticks(rotation=45)
plt.title('Age Distribution by Cat Breed')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Breeds', y='Weight')
plt.xticks(rotation=45)
plt.title('Weight Distribution by Cat Breed')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Breeds', y='length')
plt.xticks(rotation=45)
plt.title('Length Distribution by Cat Breed')
plt.tight_layout()
plt.show()
```

```python
#The relationship between weight, length, age and sleep_hours for each breed.

sns.pairplot(df, hue='Breeds', vars=['Weight', 'length', 'Age', 'Sleep_hours'])
plt.show()
```
